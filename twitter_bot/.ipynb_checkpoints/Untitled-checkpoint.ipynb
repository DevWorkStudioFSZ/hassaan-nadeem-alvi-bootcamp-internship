```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 


chrome_options = uc.ChromeOptions()
driver = uc.Chrome(options=chrome_options)

```

```python

driver.get("https://x.com/i/flow/login")
wait = WebDriverWait(driver, 15)

```

```python
email_input = driver.find_element(By.NAME,'text')
email_input.clear()
email = "EMAIL"
email_input.send_keys(email)
```

```python
next_button = driver.find_element(By.XPATH, "//button[.//span[contains(text(), 'Next')]]")
next_button.click()
```

```python
password_input = driver.find_element(By.NAME,'password')
password_input.clear()
password = "PASSWORD"
password_input.send_keys(password)
```

```python
log_in_button = driver.find_element(By.XPATH, "//button[.//span[contains(text(),'Log in')]]")
log_in_button.click()
```

```python
tweet_element = driver.find_element(By.XPATH, '//div[@data-testid="tweetText"]')
tweet_text = tweet_element.text

print(tweet_text)
```

```python

tweet_box = driver.find_element(By.XPATH, '//div[@role="textbox"]')
tweet_box.clear()
tweet_box.send_keys(tweet_text)
```

```python
post_button = driver.find_element(By.XPATH, "//button[.//span[contains(text(),'Post')]]")
post_button.click()
```
